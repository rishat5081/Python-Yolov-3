```python
import cv2
import numpy as np
net = cv2.dnn.readNet('yolov3.weights','yolov3.cfg')
classes = []
with open('coco.names','r') as f:
    classes = f.read().splitlines()
# printing the data which is loaded from the names file
#print(classes)
cap = cv2.VideoCapture('video.mp4')
person =0
truck =0
car = 0
while True:
    _,img = cap.read()
    height, width, _ = img.shape
    blob = cv2.dnn.blobFromImage(img, 1/255,  (416 , 416), (0,0,0) ,swapRB=True,crop=False)
    net.setInput(blob)
    output_layer_names = net.getUnconnectedOutLayersNames()
    layerOutput = net.forward(output_layer_names)
    x_array = np.array([])
    y_array = np.array([])
    label_array = np.array([])
    boxes = []
  
    confidences = []
    class_ids =[]
    for output in layerOutput:
        for detection in output:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0:

                center_x = int(detection[0]*width)
                center_y = int(detection[1]*height)
                # print(center_x)
                # print(center_y)
                w = int(detection[2]*width)
                h = int(detection[3]*height)


                x = int(center_x - w/2)
                y = int(center_y - h/2)

                boxes.append([x,y,w,h])
                confidences.append((float(confidence)))
                class_ids.append(class_id)
    indexes = cv2.dnn.NMSBoxes(boxes,confidences,0.5,0.4)
    font = cv2.QT_FONT_NORMAL
    colors = np.random.uniform(0,255,size=(len(boxes),3))
    for i in indexes.flatten():
        xi,yi,wi,hi = boxes[i]
        # print("x::  "+str(xi))
        # print("y::  "+str(yi))
        labelsss = str(classes[class_ids[i]])
        # print(xi)
        # print(yi)
        for line in labelsss.split('\n'): 
            if(xi > 220 and xi <350 ):
                if(yi >= 200 and yi < 202):
                    if(line == 'car'):
                        car+=1
                    print( line + " ::: Cross the Line")
                elif(yi > 204 and yi < 207 ):
                    if(line == 'car'):
                        car+=1
                    print( line + " ::: Cross the Line")

        # if(xi > 300):
        #     int_Converted = int(xi)
        #     if(yi>200):
        #         print("heellooo")
        # if(labelsss == 'person'):
        #     person+=1
        # if(labelsss == 'car'):
        #     car+=1
        # if(labelsss == 'truck'):
        #     truck+=1

    for i in indexes.flatten():
        x,y,w,h = boxes[i]
        label =str(classes[class_ids[i]])
        confidence = str(round(confidences[i],1))
        color = colors[i]
        cv2.rectangle(img,(x,y),(x+w , y+h), color, 1)
        cv2.line(img,(1000,250),(5,250),(0,0,0),2)
              
        cv2.putText(img, label + " ", (x, y+20), font, 0.5, (255,255,255),1)
        cv2.putText(img, 'Car'+ ":" + str(car), (20, 20), font, 0.8, (0,0,0),1)
        # cv2.putText(img, 'Person'+ ":" + str(person), (20, 50), font, 0.8, (0,0,0),1)
        # cv2.putText(img, 'Truck'+ ":" + str(truck), (20, 80), font, 0.8, (0,0,0),1)
        cv2.imshow('Image',img)
        key = cv2.waitKey(1)
        if key == 12:
            break
        
cap.release()
cv2.destroyAllWindows()
```
